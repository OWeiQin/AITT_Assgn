```python
# Data Cleaning
# Fill up missing Alerts/Warnings values
df['Alerts/Warnings'] = df['Alerts/Warnings'].apply(lambda x: 'yes' if x == 'Alert Triggered' else 'no')

# Fill up missing Malware Indicators values
df['Malware Indicators'] = df['Malware Indicators'].apply(lambda x: 'No Detection' if pd.isna(x) else x)

# Fill up missing Proxy Information values
df['Proxy Information'] = df['Proxy Information'].apply(lambda x: 'No Proxy' if pd.isna(x) else x)

# Fill up missing Firewall Logs values
df['Firewall Logs'] = df['Firewall Logs'].apply(lambda x: 'No Logs' if pd.isna(x) else x)

# Fill up missing IDS/IPS Alerts values
df['IDS/IPS Alerts'] = df['IDS/IPS Alerts'].apply(lambda x: 'No Alerts' if pd.isna(x) else x)
```

```python
# Check the device information column for browser info
df['Device Information'].value_counts()
```

```python
# Keep only the key browser info
df['Browser'] = df['Device Information'].str.split('/').str[0]
df['Browser']
```

```python
# Check for value count of the different browser
df['Browser'].value_counts()
```

```python
# Create column for type of device/OS
values = [
    'Windows',
    'Linux',
    'Android',
    'iPad',
    'iPod',
    'iPhone',
    'Macintosh',
]
# Create function to loop through the column to find Device/OS
def find_device_os(var):
    var = var.lower()
    for value in values:
        if value.lower() in var:
            return value
    return 'Unknown'
# Create Device/OS column and append results
df['Device/OS'] = df['Device Information'].apply(find_device_os)
df['Device/OS']
```

```python
# Check value counts for different device/OS
df['Device/OS'].value_counts()
```

```python
# Remove original Device information column
df = df.drop('Device Information', axis=1)
df.head(5).T # Check if new columns are added
```

```python
# Change timestamp to datetime for easier graph plotting
# Split the timestamp to year, month, day, dayofweek for graph
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Year'] = df['Timestamp'].dt.year
df['Month'] = df['Timestamp'].dt.month
df['Day'] = df['Timestamp'].dt.day
df['DayOfWeek'] = df['Timestamp'].dt.day_name()
df.head().T
```

```python
# Modeling
# Import libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn.linear_model import LogisticRegression
```

```python
# Data Pre-Processing
#Convert ip address to integer (remove dot in between)
import ipaddress
df['Source IP Address'] = df['Source IP Address'].apply(lambda ip: int(ipaddress.ip_address(ip)))
df['Destination IP Address'] = df['Destination IP Address'].apply(lambda ip: int(ipaddress.ip_address(ip)))
df.head().T
```

```python
# one hot encoding & dropping unimportant/non-numeric columns
df = df.drop(['Timestamp', 'Payload Data', 'User Information', 'Geo-location Data', 'Proxy Information', 'Year', 'Month', 'Day', 'DayOfWeek'], axis=1)
df = pd.get_dummies(df, columns=['Protocol', 'Packet Type', 'Traffic Type', 'Malware Indicators', 'Alerts/Warnings', 'Attack Type',
                                 'Action Taken', 'Severity Level', 'Network Segment', 'Firewall Logs', 'IDS/IPS Alerts', 'Log Source', 'Browser',
                                 'Device/OS'])
```

```python
# Change value to 0 and 1 (known pattern A = 1, known pattern B = 0)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Attack Signature'] = le.fit_transform(df['Attack Signature'])
```

```python
# Create testing and training data splits
x = df.drop(['Attack Signature'], axis=1)
y = df['Attack Signature']
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
# scaler library
scaler = StandardScaler()
```

```python
# Scale x_train and x_test data values
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
log = LogisticRegression()
```

```python
# Create a logistic regression instance
log.fit(x_train, y_train)
```

```python
# Make predictions on testing dataset
predictions = log.predict(x_test)
predictions
```

```python
# Model Evaluation
# Confusion Matrix
cm = confusion_matrix(y_test, predictions)
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(cm, index=['actual_yes','actual_no'],columns=['predicted_yes','predicted_no'])
cm_df
```

```python
# Plot confusion matrix heatmap
sns.heatmap(cm, annot=True, fmt='g', xticklabels=['1','0'], yticklabels=['1','0'])
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.xlabel('Prediciton')
plt.show()
```

```python
# Classification Report
print(classification_report(y_test, predictions))
```
